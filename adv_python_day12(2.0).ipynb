```python
#Data ---> separate(categorical,numerical)--->
#categorical data---> SimpleImputer fill---> Encode
#numerical data---> SimpleImputer fill ---> Standardize

#Apply=== Algorithm(LogisticRegression)
```

```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("covid_toy.csv")
df.head()
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler,OneHotEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
x=df.drop(columns=['has_covid'])
y=df['has_covid']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,
                                               random_state=42)
```

```python
#define the columns that need to be preprocessed
categorical_features=['gender','city']
numeric_features=['age','fever']
```

```python
#Create transformers
numeric_transformer=Pipeline(steps=[
    ('imputer',SimpleImputer(strategy='mean')),
    ('scaler',StandardScaler())
])

categorical_transformer = Pipeline(steps=[
    ('imputer',SimpleImputer(strategy='most_frequent')),
    ('onehot',OneHotEncoder(handle_unknown='ignore'))
])

#combine transformer
preprocessor=ColumnTransformer(
    transformers=[
         ('num',numeric_transformer,numeric_features),
         ('cat',categorical_transformer,categorical_features)
    ])

#Create the pipeline
clf=Pipeline(steps=[('preprocessor',preprocessor),
                    ('classifier',LogisticRegression())])

#Train the model
clf.fit(x_train,y_train)

#evaluate the model
y_pred=clf.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
```

```python
acc=accuracy_score(y_test,y_pred)
```

```python
acc
```

```python
import numpy as np
import pandas as pd
```

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,classification_report
```

```python
df=pd.read_csv("Social_Network_Ads.csv")
df.head()
```

```python
df=df.drop(columns=['User ID','Gender'])
```

```python
x=df.drop(columns=['Purchased'],axis=1)
y=df['Purchased']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,
                                               random_state=42)
pipe=Pipeline([
    ('scaler',StandardScaler()),
    ('pca',PCA(n_components=2)),
    ('classifier',RandomForestClassifier(n_estimators=100,random_state=42))
])
```

```python
# pca=>higher dimensional data ko lower m is tarah se convert karna ki uski 
# performance pr fark na pde
```

```python
pipe
```

```python
pipe.fit(x_train,y_train)
```

```python
y_pred=pipe.predict(x_test)
```

```python
acc = accuracy_score(y_test,y_pred)
print(acc)
```
