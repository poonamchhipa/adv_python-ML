```python
# knn->k nearest neighbor algo ->it is a supervised learning model that is used for both classification and regression
# but it is mainly used in classification tasks
# ->how knn work=>
# lazy learners->knn does not learn from the training data explicity means no module is build
# instead it stores entire dataset and makes the prediction at the time of testing 
# hence its called a lazy learning algo
# ->knn distance based->when we want to classify a new data point
# (1)it calculate the distance from the new point to all other data point in the training set
# (2)then it selects the k(no. of nearest neighbor) closest neighbor,
# (3)it assigns the most common class,among these neighbors to the new points
# ex.=>
# person  fever cough level(0-10) covid result
# sam     101     7      +
# rahul    98      3      -
# gita     100     6      +
# rohit    97      2      -

# now we get a new person ramesh,whose fever is 99 and cough level is 4
#     we want to predict their covid result using knn with k=3


# solution=>calculate the result we use euclidan distance between the new person and on existing distance
# step1=>formula->first points coordinates->x1 ,y1
#         second points coordinates->x2,y2
# so,distance will be=>rooot of x2-x1 ka whole square + y2-y1 ka whole square
# distance to person1=>underroot of 13->3.61
# distance to person2=>underrrot of 2->approx 1.41
# distance to person3=>underrrot of 5 ->2.24
# distance to person4=>underrrot of 8->2.83

# step2->find k nearest neighbor=>k=3
# ->sort based on distance
# person  distance results
# rahul    1.41      -
# gita     2.24      +
# rohit    2.83      -

# output of new person->negative
```

```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("covid_toy.csv")
```

```python
df.head(2)
```

```python
df=df.dropna()
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lb=LabelEncoder()
```

```python
df['gender']=lb.fit_transform(df['gender'])
df['cough']=lb.fit_transform(df['cough'])
df['city']=lb.fit_transform(df['city'])
df['has_covid']=lb.fit_transform(df['has_covid'])
```

```python
df.head(2)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=df.drop(columns=['has_covid'])
y=df['has_covid']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,
                                               random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn=KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train,y_train)
```

```python
y_pred=knn.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```

```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("insurance.csv")
```

```python
df.head(2)
```

```python
df=pd.get_dummies(df,columns=['sex','smoker','region'])
```

```python
df.head(2)
```

```python
x=df.drop(columns=['charges'])
y=df['charges']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,
                                               random_state=42)
```

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
knn=KNeighborsRegressor(n_neighbors=5)
knn.fit(x_train,y_train)
```

```python
y_pred=knn.predict(x_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,y_pred)
```
